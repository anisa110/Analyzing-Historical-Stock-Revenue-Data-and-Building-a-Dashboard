```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Target URL (Books to Scrape)
url = "https://books.toscrape.com/catalogue/page-1.html"

# Send GET request
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract book titles
titles = [title.get_text() for title in soup.find_all("h3")]

# Extract book prices
prices = [price.get_text() for price in soup.find_all("p", class_="price_color")]

# Extract stock availability
availability = [stock.get_text().strip() for stock in soup.find_all("p", class_="instock availability")]

# Create DataFrame
df = pd.DataFrame({"Title": titles, "Price": prices, "Availability": availability})

# Display first 5 records
print(df.head())
```

```python
import folium

# Create a base map centered around London
m = folium.Map(location=[51.5074, -0.1278], zoom_start=6, tiles="OpenStreetMap")

# Sample bookstore locations (latitude, longitude, name, stock)
bookstores = [
    {"name": "London Bookstore", "location": [51.5074, -0.1278], "stock": "High"},
    {"name": "Oxford Bookstore", "location": [51.7520, -1.2577], "stock": "Medium"},
    {"name": "Cambridge Bookstore", "location": [52.2053, 0.1218], "stock": "Low"},
]

# 1. Add Markers with Popups
for store in bookstores:
    folium.Marker(
        location=store["location"],
        popup=f"{store['name']}<br>Stock: {store['stock']}",
        icon=folium.Icon(color="blue", icon="book", prefix="fa")
    ).add_to(m)

# 2. Add Circle Markers (radius based on stock level)
for store in bookstores:
    radius = 3000 if store["stock"] == "High" else 2000 if store["stock"] == "Medium" else 1000
    folium.CircleMarker(
        location=store["location"],
        radius=radius/1000,  # Scale down for visibility
        color="green",
        fill=True,
        fill_opacity=0.4,
        popup=f"{store['name']} - Demand Zone"
    ).add_to(m)

# 3. Add Polyline (distribution route between bookstores)
route = [bookstores[0]["location"], bookstores[1]["location"], bookstores[2]["location"]]
folium.PolyLine(route, color="red", weight=2.5, opacity=0.8, tooltip="Distribution Route").add_to(m)

# 4. Add Additional Tile Layer (Stamen Terrain)
folium.TileLayer("Stamen Terrain").add_to(m)

# 5. Add Layer Control to toggle tiles
folium.LayerControl().add_to(m)

# Display map
m
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Sample dataset (Books to Scrape cleaned data)
data = {
    "Book_Title": ["A Light in the Attic", "Tipping the Velvet", "Soumission", "Sharp Objects", "Sapiens", 
                   "Book A", "Book B", "Book C", "Book D", "Book E"],
    "Book_Price": [51.77, 53.74, 50.10, 47.82, 54.23, 29.99, 32.50, 44.00, 60.00, 25.00],
    "Stock_Status": ["In stock", "In stock", "In stock", "In stock", "In stock", 
                     "Out of stock", "In stock", "Out of stock", "In stock", "In stock"]
}
df = pd.DataFrame(data)

# Initialize Dash app
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("Books Dashboard", style={"textAlign": "center"}),

    # Dropdown Filter for Price Range
    html.Label("Select Price Range:"),
    dcc.Dropdown(
        id="price-filter",
        options=[
            {"label": "All", "value": "all"},
            {"label": "Low (< Â£30)", "value": "low"},
            {"label": "Medium (Â£30 - Â£50)", "value": "medium"},
            {"label": "High (> Â£50)", "value": "high"},
        ],
        value="all",
        clearable=False
    ),

    # Row 1: Bar Chart & Histogram
    html.Div([
        dcc.Graph(id="bar-chart", style={"display": "inline-block", "width": "49%"}),
        dcc.Graph(id="histogram", style={"display": "inline-block", "width": "49%"})
    ]),

    # Row 2: Box Plot & Pie Chart
    html.Div([
        dcc.Graph(id="box-plot", style={"display": "inline-block", "width": "49%"}),
        dcc.Graph(id="pie-chart", style={"display": "inline-block", "width": "49%"})
    ]),

    # Scatter Plot (Interactive)
    html.Div([
        dcc.Graph(id="scatter-plot", style={"width": "100%"})
    ])
])

# Callbacks for Interactivity
@app.callback(
    [Output("bar-chart", "figure"),
     Output("histogram", "figure"),
     Output("box-plot", "figure"),
     Output("pie-chart", "figure"),
     Output("scatter-plot", "figure")],
    [Input("price-filter", "value")]
)
def update_charts(price_filter):
    # Filter data based on price selection
    if price_filter == "low":
        filtered_df = df[df["Book_Price"] < 30]
    elif price_filter == "medium":
        filtered_df = df[(df["Book_Price"] >= 30) & (df["Book_Price"] <= 50)]
    elif price_filter == "high":
        filtered_df = df[df["Book_Price"] > 50]
    else:
        filtered_df = df

    # Bar Chart - Top Books by Price
    bar_fig = px.bar(filtered_df.sort_values("Book_Price", ascending=False).head(10),
                     x="Book_Price", y="Book_Title", orientation="h",
                     title="Top Books by Price", color="Book_Price")

    # Histogram - Price Distribution
    hist_fig = px.histogram(filtered_df, x="Book_Price", nbins=10, title="Book Price Distribution", color="Book_Price")

    # Box Plot - Price Outliers
    box_fig = px.box(filtered_df, y="Book_Price", title="Price Outlier Detection")

    # Pie Chart - Stock Status
    pie_fig = px.pie(filtered_df, names="Stock_Status", title="Stock Availability Distribution")

    # Scatter Plot - Price Trends
    scatter_fig = px.scatter(filtered_df, x=filtered_df.index, y="Book_Price", 
                             hover_data=["Book_Title", "Stock_Status"], title="Price Trends Across Books")

    return bar_fig, hist_fig, box_fig, pie_fig, scatter_fig

# Run the Dash app
if __name__ == "__main__":
    app.run(debug=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Books dataset (from earlier scraping & wrangling)
data = {
    "Book_Title": ["A Light in the Attic", "Tipping the Velvet", "Soumission", "Sharp Objects", "Sapiens", 
                   "Book A", "Book B", "Book C", "Book D", "Book E"],
    "Book_Price": [51.77, 53.74, 50.10, 47.82, 54.23, 29.99, 32.50, 44.00, 60.00, 25.00],
    "Stock_Status": ["In stock", "In stock", "In stock", "In stock", "In stock", 
                     "Out of stock", "In stock", "Out of stock", "In stock", "Out of stock"]
}
df = pd.DataFrame(data)

# Encode target variable (In stock=1, Out of stock=0)
le = LabelEncoder()
df["Stock_Status"] = le.fit_transform(df["Stock_Status"])

# Features (Price) and Target (Stock_Status)
X = df[["Book_Price"]]  
y = df["Stock_Status"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================
# Model 1: Logistic Regression
# =========================
log_reg = LogisticRegression(max_iter=500, solver='liblinear')
log_reg.fit(X_train_scaled, y_train)
y_pred_lr = log_reg.predict(X_test_scaled)

# =========================
# Model 2: Random Forest
# =========================
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# Evaluation Function
def evaluate_model(y_true, y_pred, model_name):
    print(f"\nðŸ“Š {model_name} Performance:")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precision: {precision_score(y_true, y_pred):.4f}")
    print(f"Recall: {recall_score(y_true, y_pred):.4f}")
    print(f"F1 Score: {f1_score(y_true, y_pred):.4f}")
    print(f"ROC-AUC: {roc_auc_score(y_true, y_pred):.4f}")

# Evaluate both models
evaluate_model(y_test, y_pred_lr, "Logistic Regression")
evaluate_model(y_test, y_pred_rf, "Random Forest")

# =========================
# Hyperparameter Tuning (Random Forest)
# =========================
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [2, 4],
    'min_samples_split': [2, 4]
}
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, scoring='f1', n_jobs=-1)
grid_search.fit(X_train, y_train)

print("\nâœ… Best Random Forest Params:", grid_search.best_params_)
best_rf = grid_search.best_estimator_
y_pred_best_rf = best_rf.predict(X_test)

# Evaluate Tuned Random Forest
evaluate_model(y_test, y_pred_best_rf, "Tuned Random Forest")

# =========================
# Confusion Matrix Visualization
# =========================
cm = confusion_matrix(y_test, y_pred_best_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Tuned Random Forest")
plt.show()

# Classification Report
print("\nðŸ“„ Classification Report:\n", classification_report(y_test, y_pred_best_rf, target_names=le.classes_))
```
